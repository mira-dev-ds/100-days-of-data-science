Which channels appear more than once (multiple hits)?

```python
df.groupby('channel').agg(channel_count = ('channel','count')).sort_values('channel_count',ascending = False).head(10)
```

VIEWR INSIGHTS

What is the average view count?

```python
df['view_count'].mean()
```

Which video has the highest view count?

```python
df.sort_values('view_count',ascending=False).head(1)
```

Which channel has the most total views (sum of its songs)?

```python
df.groupby('channel').agg(total_views_per_channel = ('view_count','sum')).sort_values('total_views_per_channel',ascending = False).head(1)
```

Which channel has the highest followers?

```python
df.groupby('channel').agg(total_followers_per_channel = ('channel_follower_count','sum')).sort_values('total_followers_per_channel',ascending = False).head(1)
```

Duration Analysis

What is the average duration of songs?

```python
df['duration'].mean()
```

Which are the shortest and longest videos?

```python
df.sort_values('duration',ascending = False).head(1)
```

```python
df.sort_values('duration',ascending = True).head(1)
```

Tags & Categories

What are the top 10 most common tags?

```python
df.groupby('tags').agg(tag_count = ('tags','count')).sort_values('tag_count',ascending = False).head(10)
```

Which categories have the highest total views?

```python
df.groupby('categories').agg(total_views_per_category = ('view_count','sum')).sort_values('total_views_per_category',ascending = False)
```

Which category has the longest average song duration?

```python
df.groupby('categories').agg(avg_song_duration = ('duration','mean')).sort_values('avg_song_duration',ascending = False)
```

Visualize

Top 10 most viewed songs

```python
sns.barplot(x = 'view_count',y = 'title',data = df.sort_values('view_count',ascending = False).head(10))
plt.title('Top 10 most viewed songs')
plt.xlabel('View Count')
plt.ylabel(' Video Title')
plt.show()
```

Top 10 channels by total views

```python
df.groupby('channel').agg(total_views_per_channel = ('view_count','sum')).sort_values('total_views_per_channel',ascending = False).head(10).plot(kind = 'bar')
```
